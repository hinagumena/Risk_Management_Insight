```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import geopandas as gpd

import warnings

# 모든 경고 메시지 끄기
warnings.filterwarnings('ignore')

plt.rcParams['font.family'] = 'NanumBarunGothic' # 나눔바른고딕 적용하기
```

```python
y_data = pd.read_csv('/content/drive/MyDrive/서리)1차전처리데이터.csv')
x_data = pd.read_csv('/content/drive/MyDrive/서리_설명변수.csv',encoding = 'cp949')

y_data=y_data.drop('년도',axis=1)
x_data.rename(columns= {'지점명': '지점'}, inplace= True)

y_data['연']=y_data['서리끝'].str[:4].astype('int')


x_data['연'].astype('float')
x_data.dropna(inplace=True)
```

```python
region1=y_data['지점'].unique().tolist()
region2= x_data['지점'].unique().tolist()

print(set(region2)-set(region1))
```

```python
data =  pd.merge(y_data, x_data, on=['연','지점'], how='left')
```

```python
data.dropna(inplace=True)
```

```python
data.to_csv('최종_서리_데이터.csv', encoding='cp949', index=False)
```
