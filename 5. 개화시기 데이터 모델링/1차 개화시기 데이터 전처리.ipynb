```python
# 복숭아 / 배 데이터 프레임 나누기
df_복숭아_개화 = data[['지점', '년도', '복숭아 개화', '복숭아 개화(평비)']]
df_복숭아_개화
```

```python
df_복숭아_개화['복숭아 개화(평비)'].unique()
```

```python
df_복숭아_개화['복숭아 개화'].replace([' ', '결측', '관측 안됨'], np.nan, inplace=True)
df_복숭아_개화['복숭아 개화(평비)'].replace([' ', '―'], np.nan, inplace=True)
```

```python
# 결측값 확인
df_복숭아_개화.isna().sum()
```

```python
df_복숭아_개화['복숭아 개화'] = pd.to_datetime(df_복숭아_개화['복숭아 개화'])
df_복숭아_개화['복숭아 개화(평비)'] = df_복숭아_개화['복숭아 개화(평비)'].str.replace('일', '', regex=False).apply(lambda x: pd.to_numeric(x, errors='coerce'))
```

```python
# 데이터 유형 확인 후 적절한 type으로 변경
df_복숭아_개화.info()
```

```python
# 개화에 결측치가 있으면 해당 열 제거
df_복숭아_개화 = df_복숭아_개화.dropna(subset=['복숭아 개화'])
df_복숭아_개화
```

평비 결측값 채워넣기

```python
df_복숭아_개화['365'] = (df_복숭아_개화['복숭아 개화'] - pd.to_datetime(df_복숭아_개화['복숭아 개화'].dt.year, format='%Y')) + pd.Timedelta(days=1)
df_복숭아_개화['365'] = df_복숭아_개화['365'].dt.days

mean_values = df_복숭아_개화.groupby('지점')['365'].mean().reset_index()
mean_values.rename(columns={'365': '평균값'}, inplace=True)

df_복숭아_개화 = pd.merge(df_복숭아_개화, mean_values, on='지점', how='left')
df_복숭아_개화['임시(평비)'] = df_복숭아_개화['365'] - df_복숭아_개화['평균값']
df_복숭아_개화['복숭아 개화(평비)'].fillna(df_복숭아_개화['임시(평비)'], inplace=True)
```

```python
df_복숭아_개화
```

```python
df_final_복숭아개화 = df_복숭아_개화.drop(['평균값',	'임시(평비)'], axis=1)
```
