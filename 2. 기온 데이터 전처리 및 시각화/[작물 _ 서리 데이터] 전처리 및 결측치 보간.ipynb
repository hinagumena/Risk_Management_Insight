## 작물

```python
agriculture = pd.read_csv('/content/drive/MyDrive/삼성화재&Postech 데이터 리스크 /데이터/작물데이터.csv', encoding='cp949')
agriculture
```

### 컬럼 정리

```python
data_peach = agriculture.copy()
data_pear = agriculture.copy()
```

```python
# 복숭아
data_peach['과일'] = '복숭아'
data_peach.drop(['배나무', '배나무.1', '배나무.2', '배나무.3', '배나무.4', '배나무.5'],axis=1,inplace = True)
data_peach.columns = ['지점','년도','발아','발아(평비)','개화','개화(평비)','만발','만발(평비)','과일']
data_peach.drop(0, inplace = True)
```

```python
# 배
data_pear['과일'] = '배'
data_pear.drop(['복숭아', '복숭아.1', '복숭아.2', '복숭아.3', '복숭아.4', '복숭아.5'],axis=1,inplace = True)
data_pear.columns = ['지점','년도','발아','발아(평비)','개화','개화(평비)','만발','만발(평비)','과일']
data_pear.drop(0, inplace = True)
```

### 데이터프레임

```python
data_food = pd.concat([data_peach, data_pear], axis=0, ignore_index=True)
data_food
```

```python
data_food['지점'].unique() # 87개
```

```python
# 결측 데이터
(data_food['발아'].str.isspace() | data_food['발아'].eq('') | data_food['발아'].eq('결측')).sum() # 91
(data_food['개화'].str.isspace() | data_food['개화'].eq('') | data_food['개화'].eq('결측')).sum() # 95
(data_food['만발'].str.isspace() | data_food['만발'].eq('') | data_food['만발'].eq('결측')).sum() # 2765
```

```python
# 년도 정리
data_food['년도'] = pd.to_datetime(data_food['년도'], format='%Y').dt.strftime('%Y')

data_food['발아'] = pd.to_datetime(data_food['발아'], format='%Y-%m-%d', errors='coerce')
data_food['개화'] = pd.to_datetime(data_food['개화'], format='%Y-%m-%d', errors='coerce')
data_food['만발'] = pd.to_datetime(data_food['만발'], format='%Y-%m-%d', errors='coerce')
```

```python
# '일' 제거
data_food['발아(평비)'] = data_food['발아(평비)'].replace(' ', '0일')
data_food['발아(평비)'] = data_food['발아(평비)'].replace('―', '0일')

data_food['개화(평비)'] = data_food['개화(평비)'].replace(' ', '0일')
data_food['개화(평비)'] = data_food['개화(평비)'].replace('―', '0일')

data_food['만발(평비)'] = data_food['만발(평비)'].replace(' ', '0일')
data_food['만발(평비)'] = data_food['만발(평비)'].replace('―', '0일')

data_food['발아(평비)'] = data_food['발아(평비)'].str.extract(r'(-?\d+)').astype(float).astype(int)
data_food['개화(평비)'] = data_food['개화(평비)'].str.extract(r'(\d+)').astype(float).astype(int)
data_food['만발(평비)'] = data_food['만발(평비)'].str.extract(r'(\d+)').astype(float).astype(int)
```

```python
def day_of_year(date):
    if pd.isna(date):  # Check if the date is NaN
        return float('nan')  # Return NaN if it's NaN
    is_leap_year = date.year % 4 == 0 and (date.year % 100 != 0 or date.year % 400 == 0)
    if is_leap_year:
        return date.strftime('%j')
    else:
        return (date - pd.Timestamp(date.year, 1, 1)).days + 1

data_food['개화일2'] = data_food['개화'].apply(day_of_year)
```

```python
data_food['개화일'] = data_food['개화'].dt.strftime('%j').str.lstrip('0')
```

## 서리

```python
frost = pd.read_csv('/content/drive/MyDrive/삼성화재&Postech 데이터 리스크 /데이터/서리.csv', encoding='cp949')
frost
```

```python
frost.columns =['지점','년도','서리 시작','서리 시작(평비)','서리 끝','서리 끝(평비)']
frost.drop(0, inplace = True)
```

```python
frost['지점'].unique() # 87
```

```python
# 결측 데이터
(frost['서리 시작'].str.isspace() | frost['서리 시작'].eq('') | frost['서리 시작'].eq('결측')).sum() # 12
(frost['서리 끝'].str.isspace() | frost['서리 끝'].eq('') | frost['서리 끝'].eq('결측')).sum() # 62
```

```python
# 년도 정리
frost['년도'] = pd.to_datetime(frost['년도'], format='%Y').dt.strftime('%Y')

frost['서리 시작'] = pd.to_datetime(frost['서리 시작'], format='%Y-%m-%d', errors='coerce')
frost['서리 끝'] = pd.to_datetime(frost['서리 끝'], format='%Y-%m-%d', errors='coerce')
```

```python
# '일' 제거
frost['서리 시작(평비)'] = frost['서리 시작(평비)'].replace(' ', '0일')
frost['서리 시작(평비)'] = frost['서리 시작(평비)'].replace('―', '0일')

frost['서리 끝(평비)'] = frost['서리 끝(평비)'].replace(' ', '0일')
frost['서리 끝(평비)'] = frost['서리 끝(평비)'].replace('―', '0일')

frost['서리 시작(평비)'] = frost['서리 시작(평비)'].str.extract(r'(-?\d+)').astype(float).astype(int)
frost['서리 끝(평비)'] = frost['서리 끝(평비)'].str.extract(r'(-?\d+)').astype(float).astype(int)
```

```python
frost['서리 기간'] = (frost['서리 끝'] - frost['서리 시작']).dt.days
```
